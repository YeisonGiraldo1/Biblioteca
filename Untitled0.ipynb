**TRABAJANDO CON GRÁFICAS**


```python
#Leer las librerias recomendadas para trabajar con graficas en Machine Learning
import matplotlib.pyplot as plt
import seaborn as sns


#1.Ventas por ciudad
ventas_ciudad = df.groupby('Ciudad')['Cantidad'].sum().sort_values()
ventas_ciudad.plot(kind='line', color='#ff5733', title='Cantidad Total Vendida por Ciudad')
plt.ylabel('Cantidad Vendida')
plt.tight_layout()
plt.show()

```

```python

mycolor = ["red", "green" , "blue", "black", "yellow", "orange", "purple", "pink"]
ventas_categoria = df.groupby('Producto')['ingreso'].sum().sort_values()
ventas_categoria.plot(kind='bar', color=mycolor, title='Productos mas vendidos por ingreso')

plt.xlabel('Cantidad Vendida')
plt.show()
```

```python
#Graficar ventas totales por categoria (BAR)
mycolor = ["red", "green" , "blue", "black", "yellow", "orange", "purple", "pink"]
ventas_categoria = df.groupby('Categoría')['Cantidad'].sum().sort_values()
ventas_categoria.plot(kind='bar', color=mycolor, title='Cantidad Total Vendida por Categoria')

plt.xlabel('Cantidad Vendida')
plt.tight_layout()
plt.show()
```

```python
#Partipacion por categoria (PIE)
mycolor = ["red", "green" , "blue", "black", "yellow", "orange", "purple", "pink"]
ventas_categoria = df.groupby('Categoría')['Cantidad'].sum().sort_values()
ventas_categoria.plot(kind='pie', color=mycolor, autopct='%1.1f%%' , startangle = 20, title='Cantidad Total Vendida por Categoria')
plt.xlabel('Cantidad Vendida')
plt.tight_layout()
plt.show()
```

```python

import numpy as np
mycolor = ["red", "green" , "blue", "black", "yellow", "orange", "purple", "pink"]
ventas_categoria = df.groupby('Producto')['Precio'].max().sort_values()
ventas_categoria.plot(kind='line', color=mycolor, title='Productos mas costosos')
plt.ylabel('Cantidad Vendida')
plt.xlabel('')
plt.plot(ventas_categoria, marker = '*', ms = 20)
plt.tight_layout()
plt.show()
```
