Detección y Tratamiento de Datos Ausentes

Opcion 1: Eliminar los datos ❌

````python
#delete records with null values
#the function "drop" used to delete a row in specific
#the function "dropna" used to delete the null values
#in this case created new variable for store dataframe with data nulls deleted
data_weather_drop = data_weather.dropna()
data_weather_drop.head()
````

Opción 2: Aplico funciones estadisticas como la media o la mediana 🤔

````python
#sacar la media
data_weather_mean = data_weather.copy() #copia el dataset original
data_weather_mean["Wind_Speed"] = data_weather_mean["Wind_Speed"].fillna(data_weather_mean["Wind_Speed"].mean())
print(data_weather_mean)
````

````python
#sacar la mediana
data_weather_median = data_weather.copy() #copia el dataset original
data_weather_median["Wind_Speed"] = data_weather_median["Wind_Speed"].fillna(data_weather_median["Wind_Speed"].median())
print(data_weather_median)
````

**¿Cuál elegir?**
Si los datos siguen una distribución normal → Usa la media. </br>
Si hay valores atípicos o datos sesgados → Usa la mediana.

👉 Conclusión: En este caso, la media sería una buena opción para imputar valores faltantes, ya que la distribución no tiene un sesgo fuerte ni valores atípicos que la afecten significativamente.
